```python
#importing required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/home/adarsh/Desktop/tech gig project/Phising_Training_Dataset.csv')#Reading Training data set
df
```

```python
#Ploting correlation heatmap
size = max(15, len(df.columns)/2.)
f, ax = plt.subplots(figsize=(size, size))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(df.corr(),annot=False,square=True, linewidths=.7,cmap=cmap)
```

```python

y=df['Result']   #Target Variable
x=df.drop(['Result','key'],axis='columns') #Feature Matrix
```

```python
#Checking accuracy of our model trained only on 70% of training data set
# 
# 
# 
#  
# from sklearn.model_selection import train_test_split
# xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.7,random_state=100)
# from sklearn.ensemble import RandomForestClassifier
# model=RandomForestClassifier(n_estimators=100)
# model.fit(xtrain,ytrain)
# model.score(xtest,ytest)
```

```python
#Training our model on whole Training dataset
from sklearn.ensemble import RandomForestClassifier
model1=RandomForestClassifier()
model1.fit(x,y)
```

```python
#Reading our testing dataset 
df2=pd.read_csv("Phising_Testing_Dataset.csv")
df3=df2.drop("key",axis="columns")
df4=model1.predict(df3.iloc[:,:])  #Passing our dataset for prediction
df4
```

```python
#Creating dataframe from resulting array
header=["Result"]
data = pd.DataFrame(df4, columns=header)
data
```

```python
#Combining Result in testing dataframe
r = pd.concat([data,df2], axis=1)
r
```

```python
#Converting above dataframe into required csv
final=r[['key','Result']]
final.to_csv('techgig_result.csv',index=False)
```
